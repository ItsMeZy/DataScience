```python
import numpy as np
import pandas as pd

df = pd.read_csv("police_killings.csv",encoding = "ISO-8859-1")

print(df.columns,"\n\n -------------------------------------")

df
```

```python
import plotly.graph_objs as go
import plotly.offline as py

import plotly
import plotly.plotly as py

#Ora mi salvo i dati sulla popolazione totale in USA (divisi per maschi e femmine)
#L'ordine di inserimento è il seguente: 0-14 anni, 15-24 anni, 25-54 anni, 55-64 anni, 65-in poi anni
totmaleArr = [31255995, 22213952, 64528673, 20357880, 22678235]
totfemaleArr = [29919938, 21137826, 64334499, 21821976, 28376817]
#Ho a disposizione totmale e totfemale che mi memorizzano la suddivisione delle persone in base all'età (come riportato sopra)
#su tutta la popolazione americana (in base al sesso)
totmale = np.array(totmaleArr)
totfemale = np.array(totfemaleArr)

#Salvo anche le label da stampare sul chart (con lo stesso ordine in cui elaboro e memorizzo i dati)
labels = np.array(["0-14 years old", "15-24 years old", "25-54 years old", "55-64 years old", "65 or more years old"])

#Inizializzo i contatori per contare i maschi e le femmine totali (countmale e countfemale)
#ed i contatori per maschi e femmine parziali (in base all'età)
#Il tutto sulle morti per mano della polizia
male = np.zeros(len(totmale))
female = np.zeros(len(totfemale))
countmale = 0
countfemale = 0

for i in range(len(df)):
    if(df["age"].values[i]!="Unknown"):
    #Se non è riportata l'età ignoro la riga, in caso contrario mi salvo l'età presa in considerazione
        age=int(df["age"].values[i])
        if(df["gender"].values[i]=="Male"):
            countmale = countmale+1
            if((age>0)&(age<15)):
                male[0] = male[0]+1
            elif((age>14)&(age<25)):
                male[1] = male[1]+1
            elif((age>24)&(age<55)):
                male[2] = male[2]+1
            elif((age>54)&(age<65)):
                male[3] = male[3]+1
            elif(age>64):
                male[4] = male[4]+1
        elif(df["gender"].values[i]=="Female"):
            countfemale = countfemale+1
            if((age>0)&(age<15)):
                female[0] = female[0]+1
            elif((age>14)&(age<25)):
                female[1] = female[1]+1
            elif((age>24)&(age<55)):
                female[2] = female[2]+1
            elif((age>54)&(age<65)):
                female[3] = female[3]+1
            elif(age>64):
                female[4] = female[4]+1

#Avendo contato i maschi e le femmine, per una visualizzazione più pulita mi calcolo la percentuale di maschi sul totale
#e di femmine sul totale di persone in quella data fascia d'età (quante persone su 10.000.000 vengono uccise in quella fascia)
female_percentage = np.zeros(len(female))
male_percentage = np.zeros(len(male))

for i in range(len(male)):
    female_percentage[i] = female[i]/totfemale[i]*10000000
    male_percentage[i] = male[i]/totmale[i]*10000000

#A questo punto ho memorizzato i dati divisi per anni tra uomini e donne
#Plotto il grafico per i maschi
trace = go.Bar(
    x=labels,
    y=male_percentage,
    name='Men murders by police every 10 million of men'
)
data = [trace]
layout = go.Layout(
    barmode='group',
    title = "Men murders by police every 10 million of men"
)
fig = go.Figure(data=data, layout=layout)
plot_url = plotly.offline.plot(fig, validate=False, filename='Male_Age_bar.html')

#Plotto il grafico per le femmine
trace = go.Bar(
    x=labels,
    y=female_percentage,
    name='Women murders by police every 10 million of women'
)
data = [trace]
layout = go.Layout(
    barmode='group',
    title = "Women murders by police every 10 million of women"
)
fig = go.Figure(data=data, layout=layout)
plot_url = plotly.offline.plot(fig, validate=False, filename='Female_Age_bar.html')
```
